```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

train_data = pd.read_csv("/content/drive/MyDrive/trainspace.csv")
test_data = pd.read_csv("/content/drive/MyDrive/testspace.csv")
test_ids = test_data["PassengerId"]
```

```python
train_data.head(10)
```

```python
def clean(data):
  data = data.drop(["PassengerId", "Name", "Cabin", "RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"], axis=1)

  cols = ["Age"]
  for col in cols:
    data[col].fillna(data[col].mean(), inplace=True)

    #data.Embarked.fillna("U", inplace=True)

  return data
```

```python
train_data = clean(train_data)
test_data = clean(test_data)
train_data
```

```python
X = train_data.iloc[:, 0:5]
Y = train_data.iloc[:, 5]
```

```python
dummy_x = pd.get_dummies(X)
print(dummy_x.shape)
dummy_x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
sc_x = scaler.fit_transform(dummy_x)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
depth = []
for i in range(2,100,1):
  tree = DecisionTreeClassifier(criterion="entropy", max_depth = i, random_state=10)
  scores = cross_val_score(tree, sc_x, Y, cv=5)
  depth.append(scores.mean())
```

```python
final_depth = np.argmax(np.array(depth)) + 2
print(final_depth)
```

```python
final_tree = DecisionTreeClassifier(criterion="entropy", max_depth = final_depth, random_state=10)
final_tree.fit(sc_x, Y)
```

```python
dummy_test_data = pd.get_dummies(test_data)
sc_test_data_x = scaler.transform(dummy_test_data)
submission = final_tree.predict(sc_test_data_x)
```

```python
print(submission)
```

```python
submission_data = pd.DataFrame({"PassengerId": test_ids.values, "Transported": submission })
submission_data
```

```python
submission_data.to_csv("submission.csv", index=False)
```
