```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://datahub.io/core/gdp/r/gdp.csv")
data
```

這份資料記錄了許多地區不同年份的GDP，我打算看看不同地區的GDP走勢是如何的，首先看看阿拉伯國家

```python
Arab_data = data.loc[data["Country Name"] == "Arab World"]
Arab_data
```

以x軸為年分，y軸為GDP看看該區的GDP走勢

```python
Arab_x = list(Arab_data["Year"])
Arab_y = list(Arab_data["Value"])
plt.plot(Arab_x,Arab_y)
plt.title("Arab World GDP Trend", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.xlabel("Years", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.ylabel("GDP(10^12)", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.legend()
plt.show()
```

接著加入其他地區對比看看

```python

Arab_data = data.loc[data["Country Name"] == "Arab World"]
East_Asia_and_Pacific_data = data.loc[data["Country Name"] == "East Asia & Pacific"]
European_Union_data = data.loc[data["Country Name"] == "European Union"]
Latin_America_and_Caribbean_data = data.loc[data["Country Name"] == "Latin America & Caribbean"]
Middle_East_and_North_Africa_data = data.loc[data["Country Name"] == "Middle East & North Africa"]
North_America_data = data.loc[data["Country Name"] == "North America"]

Arab_x = list(Arab_data["Year"])
Arab_y = list(Arab_data["Value"])
East_Asia_and_Pacific_x = list(East_Asia_and_Pacific_data["Year"])
East_Asia_and_Pacific_y = list(East_Asia_and_Pacific_data["Value"])
European_Union_x = list(European_Union_data["Year"])
European_Union_y = list(European_Union_data["Value"])
Latin_America_and_Caribbean_x = list(Latin_America_and_Caribbean_data["Year"])
Latin_America_and_Caribbean_y = list(Latin_America_and_Caribbean_data["Value"])
Middle_East_and_North_Africa_x = list(Middle_East_and_North_Africa_data["Year"])
Middle_East_and_North_Africa_y = list(Middle_East_and_North_Africa_data["Value"])
North_America_x = list(North_America_data["Year"])
North_America_y = list(North_America_data["Value"])

plt.plot(Arab_x, Arab_y, color = 'red', label = 'Arab' )
plt.plot(East_Asia_and_Pacific_x, East_Asia_and_Pacific_y, color = 'yellow', label = 'East Asia & Pacific' )
plt.plot(European_Union_x, European_Union_y, color = 'blue', label = 'Europe' )
plt.plot(Latin_America_and_Caribbean_x, Latin_America_and_Caribbean_y, color = 'green', label = 'Latin America & Caribean' )
plt.plot(Middle_East_and_North_Africa_x, Middle_East_and_North_Africa_y, color = 'purple', label = 'Middle East &North Africa' )
plt.plot(North_America_x, North_America_y, color = 'orange', label = 'North America' )

plt.title("Regional GDP Trend", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.xlabel("Years", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.ylabel("GDP(10^13)", fontdict={'fontname': 'Arial Black', 'fontsize': 16})
plt.legend()
plt.show()    



```

可以看到這幾個地區在50年間的GDP大致是逐年上升，但歐洲、拉美及加勒比、中東及北非、阿拉伯地區在近十年有持平及減緩的現象，而亞太地區及北美地區則繼續向上升。
